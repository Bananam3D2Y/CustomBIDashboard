```python
import pandas as pd
import numpy as np
```

```python
file_path = "TXT Files/MainData.txt"
```

```python
def read_to_df(file_path):
    df = pd.read_csv(
        file_path,
        sep=r"\|\|",
        engine="python",
        encoding="cp1252",
        dtype=str
    )

    # Clean quotes and whitespace
    df = df.apply(lambda col: col.str.strip().str.strip('"'))
    df.columns = df.columns.str.strip().str.strip('"')

    # Infer types
    df = df.apply(pd.to_numeric, errors="ignore")
    return df
```

```python
# Read single txt file into a df
MainData = read_to_df("TXT Files/MainData.txt")
Stores = read_to_df("TXT Files/Stores.txt")
```

```python
# Get all unique store ids which are not negative
df1 = MainData['StoreID']
df2 = Stores[['StoreID','StoreName']]
df3 = pd.merge(df1, df2, on="StoreID", how="left")
df3.head()
```
